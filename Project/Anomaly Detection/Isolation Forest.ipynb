```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Create some sample data (1D array for simplicity)
data = np.random.normal(scale=0.5, size=1000)

# Reshape the data to fit the model input format
data = data.reshape(-1, 1)

# Create an Isolation Forest model
clf = IsolationForest(contamination=0.1) # Contamination is the proportion of Anomaly. basically 0.05 mean 5%, 0.1 mean 10% etc. 

# Fit the model and predict the anomaly
predictions = clf.fit_predict(data)

# Anomalies are labeled as -1, normal points are labled as 1
anomaly_indices = np.where(predictions == -1)[0]
```

```python
plt.plot(range(len(data)), data)
# Highlight the anomalies
# scatter plot is used to plot data points on horizontal and vertical axis in the graph 
plt.scatter(anomaly_indices, data[anomaly_indices], c='r')
anomaly = []
anomaly.append(anomaly_indices)
print(anomaly)
plt.show()
```
